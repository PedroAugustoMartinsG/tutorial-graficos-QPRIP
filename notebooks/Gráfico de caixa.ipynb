**Passo 1: Preparar os dados**

```python
# Bibliotecas utilizadas
import pandas as pd
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Carregar base de dados
df = pd.read_csv('exemplos_QPRIP.csv')
```

```python
# Este tipo de gráfico precisa, por exemplo, da posição institucional do respondente
# Para manter o anonimato, esses dados são sintéticos, ou seja, falsos

# Posições institucionais pré-definidas
posicoes = [
    'Aluno(a) de graduação',
    'Aluno(a) de pós-graduação',
    'Pós-doutorando(a)',
    'Servidor(a) docente',
    'Servidor(a) técnico(a)-administrativo(a)'
]

# Atribui uma posição aleatória para cada registro
df['Qual a sua posição principal na USP hoje?'] = np.random.choice(posicoes, size=len(df))

# Atribuir idade aleatória entre 17 e 70
df['Idade'] = np.random.randint(17, 71, size=len(df))
```

```python
# Definir coluna alvo
coluna_alvo = 'Idade'

# Definir título do eixo x
titulo_eixo_x = 'Qual a sua posição principal na USP hoje?'

# Definir título do eixo y
titulo_eixo_y = 'Qual é a sua idade?'

# Ordem desejada das categorias
ordem_categorias = [
    "Aluno(a) de graduação",
    "Aluno(a) de pós-graduação",
    "Pós-doutorando(a)",
    "Servidor(a) docente",
    "Servidor(a) técnico(a)-administrativo(a)"
]
```

---

**Parte 2: Gerar gráfico de barras horizontais agrupadas**

```python
# Criar o gráfico de boxplots
plt.figure(figsize=(11, 6))  # Define o tamanho da figura (largura=11, altura=6 polegadas)
sns.set_theme(style="darkgrid")  # Aplica tema escuro com grid de fundo

# Cria o boxplot com dados e ordem definida das categorias
sns.boxplot(
    x='Qual a sua posição principal na USP hoje?',  # Coluna usada no eixo X
    y='Idade',  # Coluna usada no eixo Y
    data=df.reset_index(drop=True),  # Usa os dados do DataFrame com índice resetado
    palette=sns.color_palette("tab10", 5),  # Define a paleta de cores com 5 categorias
    order=ordem_categorias  # Define a ordem personalizada das categorias no eixo X
)

# Define o rótulo do eixo X
plt.xlabel('Qual a sua posição principal na USP hoje?', fontsize=11)

# Define o rótulo do eixo Y
plt.ylabel('Qual é a sua idade?', fontsize=11)

# Define o tamanho da fonte dos rótulos do eixo X
plt.xticks(fontsize=9)

# Ajusta o layout para evitar cortes ou sobreposição de elementos
plt.tight_layout()

# Salvar em PDF com alta resolução e ajustes de borda (acessível para leitores de tela)
plt.savefig('idade.pdf', bbox_inches='tight', dpi=300)

# Exibe o gráfico na tela
plt.show()
```
